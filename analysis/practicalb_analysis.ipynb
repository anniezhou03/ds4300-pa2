```python
import pandas as pd
import matplotlib.pyplot as plt
faiss = pd.read_csv('data.csv')
faiss.head()
```

```python
print('time by chunk size')
chunk300 = faiss[faiss['chunk size'] == 300]
avg_time_chunk300 = chunk300['time (seconds)'].mean()
print(avg_time_chunk300)

chunk500 = faiss[faiss['chunk size'] == 500]
avg_time_chunk500 = chunk500['time (seconds)'].mean()
print(avg_time_chunk500)

chunk1000 = faiss[faiss['chunk size'] == 1000]
avg_time_chunk1000 = chunk1000['time (seconds)'].mean()
print(avg_time_chunk1000)

print('time by overlap')
overlap0 = faiss[faiss['overlap'] == 0]
avg_time_overlap0 = overlap0['time (seconds)'].mean()
print(avg_time_overlap0)

overlap50 = faiss[faiss['overlap'] == 50]
avg_time_overlap50 = overlap50['time (seconds)'].mean()
print(avg_time_overlap50)

overlap100 = faiss[faiss['overlap'] == 100]
avg_time_overlap100 = overlap100['time (seconds)'].mean()
print(avg_time_overlap100)
```

```python
print('memory by chunk size')
avg_mem_chunk300 = chunk300['memory (MB)'].mean()
print(avg_mem_chunk300)

avg_mem_chunk500 = chunk500['memory (MB)'].mean()
print(avg_mem_chunk500)

avg_mem_chunk1000 = chunk1000['memory (MB)'].mean()
print(avg_mem_chunk1000)

print('memory by overlap')
avg_mem_overlap0 = overlap0['memory (MB)'].mean()
print(avg_mem_overlap0)

avg_mem_overlap50 = overlap50['memory (MB)'].mean()
print(avg_mem_overlap50)

avg_mem_overlap100 = overlap100['memory (MB)'].mean()
print(avg_mem_overlap100)
```

```python
em_df = faiss.groupby("embedding model")[["time (seconds)", "memory (MB)"]].mean()

plt.figure(figsize=(8, 5))
plt.plot(em_df.index, em_df['time (seconds)'], label='Time (Seconds)', marker='o', color='blue')
plt.plot(em_df.index, em_df['memory (MB)'], label='Memory (MB)', marker='x', color='green')

plt.title('Average Time and Memory by Embedding Model with FAISS DB')
plt.xlabel('Embedding Model')
plt.ylabel('Time and Memory')

plt.legend(title="Metrics")
plt.show()
```

```python
em_df = faiss.groupby("chunk size")[["time (seconds)", "memory (MB)"]].mean()

plt.figure(figsize=(8, 5))
plt.plot(em_df.index, em_df['time (seconds)'], label='Time (Seconds)', marker='o', color='blue')
plt.plot(em_df.index, em_df['memory (MB)'], label='Memory (MB)', marker='x', color='green')

plt.title('Average Time and Memory by Chunk Size with FAISS DB')
plt.xlabel('Chunk Size')
plt.ylabel('Time and Memory')

plt.legend(title="Metrics")
plt.show()
```

```python
em_df = faiss.groupby("overlap")[["time (seconds)", "memory (MB)"]].mean()

plt.figure(figsize=(8, 5))
plt.plot(em_df.index, em_df['time (seconds)'], label='Time (Seconds)', marker='o', color='blue')
plt.plot(em_df.index, em_df['memory (MB)'], label='Memory (MB)', marker='x', color='green')

plt.title('Average Time and Memory by Chunk Overlap with FAISS DB')
plt.xlabel('Chunk Overlap')
plt.ylabel('Time and Memory')

plt.legend(title="Metrics")
plt.show()
```
